```python
# https://thecleverprogrammer.com/2021/04/21/uber-trips-analysis-using-python/

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
data = pd.read_csv("uber.csv")
data["Date/Time"] = pd.to_datetime(data["Date/Time"])  # Convert to datetime

# Extract day, weekday, and hour
data["Day"] = data["Date/Time"].dt.day
data["Weekday"] = data["Date/Time"].dt.weekday
data["Hour"] = data["Date/Time"].dt.hour

# Display the first few rows of the data
print(data.head())

# Set the figure size
sns.set(rc={'figure.figsize':(12, 10)})

# Plot the distribution of days
plt.subplot(2, 1, 1)
sns.distplot(data["Day"], kde=False, bins=31)  # Assuming maximum 31 days in a month
plt.title('Distribution of Days')

# Plot the distribution of hours
plt.subplot(2, 1, 2)
sns.distplot(data["Hour"], kde=False, bins=24)
plt.title('Distribution of Hours')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
